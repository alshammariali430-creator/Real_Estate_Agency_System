```python
divorce.loc[divorce[" Husband's age at divorce"] >= 39 , "new age"] = 1
divorce.loc[divorce[" Husband's age at divorce"] < 39 , "new age"] = 2
divorce.loc[divorce[" Wife's age at divorce"] >= 34 , "new wife age"] = 3
divorce.loc[divorce[" Wife's age at divorce"] < 34 , "new wife age"] = 4
divorce.loc[divorce["Year"]==2017, "new year"]=2017
divorce.loc[divorce["Year"]==2018, "new year"]=2018
divorce.loc[divorce["Year"]==2019, "new year"]=2019
divorce
```

```python
divorce.reset_index(inplace=True)
divorce.set_index("new year", inplace=True)
divorce[" Husband's age at divorce"].plot.bar(title="age", color=["Red","green","Blue"])
```
